```python
#Unsupervised learning is a type of machine learning where the model is trained on unlabeled data to find patterns and relationships in the data. 
# It is often used for clustering, dimensionality reduction, and anomaly detection. 
# In unsupervised learning, the algorithm tries to learn the underlying structure of the data without any prior knowledge of the labels or categories.
#It doesnt have predefined output labels, and the model is expected to identify patterns and groupings in the data on its own.

#Real worls applications:
#Clustering, anomaly detection. and dimentionality reduction
'''
-Customer segmentation  : Grouping customers based on purchasing behaviour
-Anomaly detection : Identifying unusual patterns in data, such as fraud detection
-Dimensionality reduction : Reducing the number of features in a dataset while preserving important information, such as in image compression
-Recommender systems : Suggesting products or content to users based on their preferences and behavior
-image and video analysis : Grouping similar images or videos together,for organization or retrieval purposes
-Document clustering : Organising large collections of documents into meaningful clusters for easier navigation and search
-Medical diagnosis : Identifying patterns in medical data to assist in diagnosis and treatment planning
-Social Network analysis : Identifying communities and relationships within social networks

'''
```

```python
#reinforcement learning is a type of machine learning where an agent learns to make decisions by interacting with an environment.
import random
#Environment
goal = 5
position = 0

#output
print('Robot is at position:', position)

while position != goal:
    #Action, random action
    action = random.choice(['Left', 'Right'])  # Move left (-1) or right (+1)

    if action == 'Right':
        position += 1
    else:
        position = max(0, position - 1)  # Move left, but not below 0

    print('Robot is at position:', position)
    #Reward
    if position == goal:
        reward = 100
    else:
        reward = -1

#Output of the action of the robot
    print(f'Action: {action:5}, Position: {position}, Reward: {reward}')

print('\n Robot reached the goal!')
```
