```python
# =====================================================
# Image Clustering Example using K-Means
# Example: Grouping Images Automatically
# =====================================================

# Import necessary libraries
import numpy as np
from sklearn.cluster import KMeans
```

```python
# ----------------------------------------------------------
# Image Names
# (Used only for displaying results)
# ----------------------------------------------------------

image_names = [

    "dog_1.jpg","dog_2.jpg","dog_3.jpg",
    "dog_4.jpg","dog_5.jpg","dog_6.jpg",

    "cat_1.jpg","cat_2.jpg","cat_3.jpg",
    "cat_4.jpg","cat_5.jpg","cat_6.jpg",

    "car_1.jpg","car_2.jpg","car_3.jpg",
    "car_4.jpg","car_5.jpg","car_6.jpg",

    "mountain_1.jpg","mountain_2.jpg","mountain_3.jpg",
    "mountain_4.jpg","mountain_5.jpg","mountain_6.jpg",

    "unknown_1.jpg",
    "unknown_2.jpg",
    "unknown_3.jpg",
    "unknown_4.jpg",
    "unknown_5.jpg",
    "unknown_6.jpg"

]
```

```python
# ----------------------------------------------------------
# Sample Image Features
#
# Suppose every image is converted into two numerical
# features:
#
# Feature 1 = Color
# Feature 2 = Texture
#
# These values are made up ONLY for demonstration.
# ----------------------------------------------------------

image_features = np.array([

# Dogs
[12,18],[14,20],[13,17],[15,19],[11,18],[16,20],

# Cats
[60,65],[62,67],[59,64],[63,66],[61,63],[60,68],

# Cars
[120,20],[118,18],[122,22],[119,19],[121,21],[123,20],

# Mountains
[200,180],[198,182],[201,179],
[203,181],[199,178],[202,183],

# Unknown Images
[250,50],     # Flower
[180,20],     # Airplane
[170,100],    # Bicycle
[90,150],     # City
[230,90],     # Burger
[180,200]     # Beach

])
```

```python
# Create KMeans Model
# ----------------------------------------------------------

kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

# Train the model

kmeans.fit(image_features)

# Cluster assignments

clusters = kmeans.labels_
```

```python
# Display Results
# ----------------------------------------------------------

print("="*50)
print("IMAGE CLUSTERING RESULTS")
print("="*50)

for image, cluster in zip(image_names, clusters):

    print(f"{image:18} ---> Cluster {cluster}")

# ----------------------------------------------------------
# Cluster Centers
# ----------------------------------------------------------

print("\nCluster Centers")
print("-"*40)

print(kmeans.cluster_centers_)
```
